# Energy Optimization Dashboard

## Project Overview
This notebook simulates 365 days of operating data for a distillation column and uses a machine learning model to optimize energy consumption.

## Goals
Understand the full distillation system
Write code to simulate a distillation column running for 365 days using NumPy and Pandas
Store the data in SQLite database
Visualize the data using Pandas, Matplotlib, and Seaborn
Train a regression model on the data using Scikit-learn, Pandas/NumPy, and Matplotlib/Seaborn
Optimize the trained model to find target operation conditions for reduced energy consumption using Scipy, NumPy, and Matplotlib

## Why this matters
Energy is the largest operating cost in chemical plants using a distillation system. This tool identifies optimal operating conditions to minimize energy use while also respecting realistic process contraints.
